```python
import ast
import numpy as np

# Final conversion of all strings into lists

final_raw_data = []
approved_points = 0

#Getting all files
for x in range (0,content_count):
    try:
        final_raw_data.append(ast.literal_eval(collected_information[0][x]))
        approved_points += 1
    except ValueError:
        pass
    
extracted_data = []
for x in range(0, approved_points):
    extracted_data.append(final_raw_data[x][0]) #States
    for y in range(1,2):
        extracted_data[x].append(final_raw_data[x][-1]) #Reward
        extracted_data[x].append(final_raw_data[x][y])  #Action
            
        
    print(extracted_data[x])
```

```python
training_data = []
testing_data = []
test_sample_value = 16

for x in range (0,approved_points):
    if x < approved_points - test_sample_value:
        training_data.append(extracted_data[x])
    else:
        testing_data.append(extracted_data[x])
        
training_data = numpy.array(training_data)
testing_data = numpy.array(testing_data)

requesting_data = []
tracking = 0
for point in testing_data:
    requesting_data.append(np.append(point[0:7],np.nan))

    
print(requesting_data)
```

```python
from spn.algorithms.LearningWrappers import learn_parametric, learn_classifier
from spn.structure.leaves.parametric.Parametric import Categorical, Gaussian
from spn.structure.Base import Context
spn_classification = learn_classifier(training_data,
                       Context(parametric_types=[Gaussian, Gaussian, Gaussian, Gaussian, Gaussian, Gaussian, Gaussian, Categorical]).add_domains(training_data),
                       learn_parametric, 7)
```

```python
from spn.algorithms.MPE import mpe
results = mpe(spn_classification, requesting_data)
```

```python
print(results)
```

```python
correct = 0

for x in range(0, test_sample_value):
    if testing_data[x][-1] == results[x][-1]:
        print(results[x])
        correct += 1
        
print("Accuracy: " + str(correct/test_sample_value))
```
